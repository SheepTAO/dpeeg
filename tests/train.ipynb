```python
import os
import sys

BOOTPATH = os.path.dirname(sys.path[0])
sys.path.insert(0, BOOTPATH)
import dpeeg
print(dpeeg.__version__)
```

```python
from dpeeg.data import transforms, datasets, preprocessing

trans = transforms.ComposeTransforms(
    transforms.Normalization(),
    transforms.Unsqueeze(),
)

preps = preprocessing.ComposePreprocess(
    preprocessing.Filter(lfreq=4, hfreq=40, method='iir')
)

data = datasets.BCICIV2A(subjects=[1], tmin=0.5, tmax=3.5, transforms=trans, preprocess=preps)
```

```python
from dpeeg.models import EEGNet

net = EEGNet(nCh=22, nTime=750)

trainer = dpeeg.Train(net, nGPU=1, dataSize=data.size, optimizer='Adam')
exp = dpeeg.Holdout(trainer, outFolder='out')
res = exp.run(data, data.clsName)
```
