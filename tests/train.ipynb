```python
import os
import sys

BOOTPATH = os.path.dirname(sys.path[0])
sys.path.insert(0, BOOTPATH)
import dpeeg
print(dpeeg.__version__)
```

```python
from dpeeg.data import transforms, datasets

trans = transforms.ComposeTransforms(
    transforms.Normalization(),
    transforms.Unsqueeze(),
)

data = datasets.BCICIV2A(tmin=0, tmax=4, transforms=trans)
```

```python
from dpeeg.models import EEGNet

net = EEGNet(nCh=22, nTime=1000)

trainer = dpeeg.Train(net, nGPU=1, data_size=data.size, optimizer='Adam')
exp = dpeeg.KFold(trainer, k=5, out_folder='out')
res = exp.run(data, data.cls_name)
```
